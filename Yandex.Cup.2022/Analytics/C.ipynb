```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('./dataset.tsv', sep='\t', names=['Day', 'A/B', 'idx', 'value'], dtype={'A/B': 'category'})
df['A/B'].cat.rename_categories({1: 'A', 2: 'B'})
df
```

```python
fig, axis = plt.subplots(4, 4)
feat = df[df['idx'] == 34]
for i, ax in enumerate(axis.flat):
    ax.hist(feat[feat['Day'] <= i]['value'] * 15 - i)
    ax.hist(feat[feat['Day'] > i]['value'] * i + 1)
```

```python
for i in range(1500):
    feat = df[df['idx'] == i]
    g = feat.groupby('Day')
    print(g['value'].min(), g['value'].max())
```
