```python
import math
from itertools import combinations, chain
```

```python
def x(n, weights):
    w_s = set(weights)

    if n > sum(weights):
        print('No')
        return

    res = set()
    z = len(weights)
    for i in range(z // 2 + 1):
        s = 0
        for g in combinations(range(z), i):
            s_g = set(g)
            for j in range(z):
                if j not in s_g:
                    s -= weights[j]
                else:
                    s += weights[j]
                res.add(abs(s))
    return res
```

```python
x(94, [34, 7, 4, 25, 30, 27, 39])
```
