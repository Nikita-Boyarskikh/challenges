```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy import integrate
from functools import partial
from tqdm.notebook import tqdm
%matplotlib inline
```

```python
inputs = [
    (150, 80),
    (310, 60),
    (450, 70),
    (550, 70),
    (730, 85),
    (900, 90),
]

data = np.arange(1,1000,1)
def total(data):
    return np.max(np.array([norm.pdf(data, loc=loc, scale=scale) for loc, scale in inputs]))
total = np.vectorize(total)
```

```python
plt.plot(data, total(data))
plt.axvline(x=156, color='green')
plt.axvline(x=356, color='green')
plt.axvline(x=400, color='yellow')
plt.axvline(x=600, color='yellow')
```

```python
for i in range(398, 403):
    val, x = integrate.quad_vec(total, i, i + 200)
    print(i, val)
```
